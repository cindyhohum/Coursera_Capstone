### 2. Second, I'll import data and clean it up

```python
# import the table data from Wikipedia as a pandas dataframe and check row and column size
df_test = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df_test.shape
```

```python
# remove instances of 'Not assigned'
df_toronto = df_test[df_test.Borough != 'Not assigned'].reset_index(drop=True)

# one row per postal code, comma delineated neighborhoods
df_final = df_toronto.groupby(['Postal Code','Borough'], as_index=False).agg(lambda x: ','.join(x))
mask = df_final['Neighbourhood'] == "Not assigned"
df_final.loc[mask, 'Neighbourhood'] = df_final.loc[mask, 'Borough']

# check the results
df_final
```

```python
# final check
df_final.shape
```

### 3. Third, I'll translate each neighborhood into latitude and longitude coordinates

```python
# import lat long data and merge dataframes
df_latlong = pd.read_csv('Geospatial_Coordinates.csv')
df_latlong = pd.merge(df_final, df_latlong, on = 'Postal Code')

# check the results
df_latlong
```

```python
# filter data to only include Downtown Toronto in the Borough column
df_dt = df_latlong.drop(df_latlong.index[df_latlong.Borough != 'Downtown Toronto'])
df_dt
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# convert json data into pandas dataframe
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
# how much data is being returned?
print('{} venues were returned by Foursquare.'.format(nearby_venues.shape[0]))
```

```python
# need more data for clustering, open the get results parameters
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
nearby_venues
```

```python
# define new dataframe to connect data and venues into one frame
toronto_venues = getNearbyVenues(names=df_dt['Neighbourhood'],
                                  latitudes=df_dt['Latitude'],
                                  longitudes=df_dt['Longitude']
                                  )
toronto_venues
```

```python
# one hot encoding to analyze each neighborhood
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighbourhood'] = toronto_venues['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
# check size
toronto_onehot.shape
```

```python
# group by neighborhood with mean and frequency of occurence in each category 
toronto_grouped = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
toronto_grouped
```

```python
# compare new size
toronto_grouped.shape
```

```python
# let's look at the top 5 venues for each neighborhood
num_top_venues = 5

for hood in toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# sort venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# pass data into new pandas dataframe
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
toronto_venues_sorted = pd.DataFrame(columns=columns)
toronto_venues_sorted['Neighbourhood'] = toronto_grouped['Neighbourhood']

for ind in np.arange(toronto_grouped.shape[0]):
    toronto_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

toronto_venues_sorted.head()
```

### 6. Now I can cluster the results and create a final, color-coded map

```python
# run k-means clustering
kcluster = 5
toronto_grouped_clustering = toronto_grouped.drop('Neighbourhood', 1)
kmeans = KMeans(n_clusters = kcluster, random_state = 0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:20]
```

```python
# add clustering label array into new column
toronto_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

# merge toronto_venues with toronto_venues_sorted to add latitude/longitude data back in
toronto_merged = toronto_venues.join(toronto_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

toronto_merged.head()
```

```python
# create map
map_clusters = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# set color array for the clusters
x = np.arange(kcluster)
ys = [i + x + (i*x)**2 for i in range(kcluster)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = []
for i in colors_array:
    rainbow.__iadd__([myColors.rgb2hex(i)])

# test output
print(rainbow)
```

```python
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Venue Latitude'], toronto_merged['Venue Longitude'], toronto_merged['Venue'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster]
        ).add_to(map_clusters)
       
map_clusters
```
